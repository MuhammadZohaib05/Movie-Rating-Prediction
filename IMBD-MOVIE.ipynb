23. Classify Movies by Rating

```python
def rating_category(x):
    if x >= 8: return "Excellent"
    elif x >= 6: return "Good"
    else: return "Average"

data['Rating_Category'] = data['Rating'].apply(rating_category)
data['Rating_Category'].value_counts()
```

26. Number of Films per Genre

```python
from collections import Counter
genre_count = Counter([g for s in data['Genre'].dropna().str.split(',') for g in s])
pd.DataFrame(genre_count.items(), columns=['Genre','Count']).sort_values(by='Count', ascending=False)
```

27. Correlation Heatmap

```python
plt.figure(figsize=(10,6))
sns.heatmap(data.select_dtypes(include=['float64','int64']).corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Numeric Features")
plt.show()
```

```python
# Keep only numeric columns
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Correlation matrix
corr = numeric_data.corr()

# Sort correlations with Revenue
revenue_corr = corr['Revenue (Millions)'].sort_values(ascending=False)
print(revenue_corr)
```
